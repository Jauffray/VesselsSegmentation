# A Simple Model

```python
# import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Define the convolutional block
# First we define the convolutional block so we do not need to re-write it every time
class conv_block(nn.Module):
    """
    Define the convolutional - batch norm - relu block to avoid re-writing it
    every time
    """

    def __init__(self, in_size, out_size, kernel_size=3, padding=1, stride=1, apply_relu=True):
        super(conv_block, self).__init__()
        self.conv = nn.Conv2d(in_size, out_size, kernel_size,
                              padding=padding, stride=stride)
        self.bn = nn.BatchNorm2d(out_size)
        self.relu = nn.ReLU(inplace=True)
        self.apply_relu = apply_relu
    
    def forward(self, x):
        if self.apply_relu==True:
            return self.relu(self.bn(self.conv(x)))
        else:
            return self.conv(x)
```

```python
# Define the network
# Then we define the network keeping in mind that down_1 output size should match with
# the scaled output of up_2, while down_2 size should match with the scaled output of the middle 
# convolution of the u-net model.
class small_UNET(nn.Module):
    """
    Define UNET model that accepts a 256 input and mostly uses 3x3 kernels
    with stride and padding = 1. It reduces the size of the image to 8x8 pixels
    ** It might not work if the input 'x' is not a square.
    """

    def __init__(self):
        super(small_UNET, self).__init__()

        self.down_1 = nn.Sequential(
            conv_block(3, 16),
            conv_block(16, 32, stride=2, padding=1))

        self.down_2 = nn.Sequential(
            conv_block(32, 64),
            conv_block(64, 128))

        self.middle = conv_block(128, 128, kernel_size=1, padding=0)

        self.up_2 = nn.Sequential(
            conv_block(256, 128),
            conv_block(128, 32))

        self.up_1 = nn.Sequential(
            conv_block(64, 64),
            conv_block(64, 32))

        self.output = nn.Sequential(
            conv_block(32, 16),
            conv_block(16, 1, kernel_size=1, padding=0, apply_relu=False))

    def forward(self, x):
#         print(x.size())

        down1 = self.down_1(x)
        out = F.max_pool2d(down1, kernel_size=2, stride=2)
#         print(out.size())
        
        down2 = self.down_2(out)
        out = F.max_pool2d(down2, kernel_size=2, stride=2)
#         print(out.size())
        
        out = self.middle(out)
#         print(out.size())
        
        out = F.upsample(out, scale_factor=2)
        out = torch.cat([down2, out], 1) ### SKIP CONNECTION
        out = self.up_2(out)
#         print(out.size())
        
        out = F.upsample(out, scale_factor=2)
        out = torch.cat([down1, out], 1) ### SKIP CONNECTION
        out = self.up_1(out)
#         print(out.size())
        
        out = F.upsample(out, scale_factor=2)
#         print(out.size())
        
        final = F.sigmoid(self.output(out))
#         print(final.size())
        
        return final
```

```python
# Encapsulate the Binary Cross Entropy loss function
class BCELoss2d(nn.Module):
    def __init__(self, weight=None, size_average=True):
        super(BCELoss2d, self).__init__()
        self.bce_loss = nn.BCELoss(weight, size_average)

    def forward(self, probs, labels):
        probs_flat = probs.view(-1)
        labels_flat = labels.view(-1)
        return self.bce_loss(probs_flat, labels_flat)
```

## Training one Epoch

```python
import torch.optim as optim

model = small_UNET().cuda()

criterion = BCELoss2d()

optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Set the model to trainable:
model.train()
# extract a batch of images and labels
i, (images, labels, masks) = next(enumerate(train_loader))
# move them to the gpu:
images = Variable(images.cuda())
labels = Variable(labels.cuda())
# masks = Variable(masks.cuda())

# before using the model to generate predictions remember to zero_grad the optimizer
optimizer.zero_grad()
# compute predictions with the model on its current state
outputs = model(images)
# compute corresponding loss
loss = criterion(outputs, labels)
# compute gradient of loss function wrt model weights
loss.backward()
# perform one step of gradient descent
optimizer.step()
```

```python
nb_epochs = 2
for epoch in range(nb_epochs):
    # Train on the dataset batch-by-batch
    for i, (images, labels, masks) in enumerate(train_loader):
        # Train on this batch
        print('Batch: ', i+1, '/', len(train_loader))
        
    print('Finished Epoch ', epoch+1, '/', nb_epochs)
```

```python
def train_epoch(train_loader, optimizer, model, loss, epoch, nb_epochs):
    for i, (images, labels, masks) in enumerate(train_loader):
#         print('Batch: ', i+1, '/', len(train_loader))

        # move them to the gpu:
        images = Variable(images.cuda())
        labels = Variable(labels.cuda())
        # before using the model to generate predictions remember to zero_grad the optimizer
        optimizer.zero_grad()
        # compute predictions with the model on its current state
        outputs = model(images)
        # compute corresponding loss
        loss = criterion(outputs, labels)
        # compute gradient of loss function wrt model weights
        loss.backward()
        # perform one step of gradient descent
        optimizer.step()

    print('Finished Epoch ', epoch+1, '/', nb_epochs)
    return optimizer, model, loss
```

```python
model = small_UNET().cuda()

criterion = BCELoss2d()

optimizer = optim.Adam(model.parameters(), lr=0.001)
# Set the model to trainable:
model.train()

optimizer, model, loss = train_epoch(train_loader, optimizer, model, loss, epoch = 0, nb_epochs = 3)
print(to_np(loss))
```
