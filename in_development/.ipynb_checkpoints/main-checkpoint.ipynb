```python
from PIL import Image
import os.path as osp
from torchvision import transforms as tr
from torch.utils.data.dataset import Dataset
from torch.utils.data import DataLoader
```

```python
class VesselsDataset(Dataset):
    def __init__(self, img_dir, gt_dir, transforms=None):
        self.img_dir = img_dir
        self.gt_dir = gt_dir
        self.img_names = sorted(os.listdir(self.img_dir))
        self.gt_names = sorted(os.listdir(self.gt_dir))
        self.transforms = transforms

    def __getitem__(self, index):
        img = Image.open(osp.join(self.img_dir, self.img_names[index]))
        gt = Image.open(osp.join(self.gt_dir, self.gt_names[index]))
        if self.transforms is not None:
            img = self.transforms(img)
            gt = self.transforms(gt)
        return img, gt

    def __len__(self):
        return len(self.img_names)
```

```python
resz = tr.Resize([512,512])
to_tensor = tr.ToTensor()

my_dataset = VesselsDataset(PATH_IMS, PATH_GT, transforms = tr.Compose([resz, to_tensor]))
my_val_dataset = VesselsDataset(PATH_VAL_IMS, PATH_VAL_GT, transforms = tr.Compose([resz, to_tensor]))
```

```python
my_loader = DataLoader(dataset=my_dataset,
                        batch_size=2, num_workers=0, shuffle=True)
my_val_loader = DataLoader(dataset=my_val_dataset,
                        batch_size=1, num_workers=0, shuffle=True)
```

# U-Net Class

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class double_conv(nn.Module):
    '''(conv => BN => ReLU) * 2'''
    def __init__(self, in_ch, out_ch):
        super(double_conv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1), #become a sum of features maps
            nn.BatchNorm2d(out_ch), #normalizing values, even in hidden layers
            nn.ReLU(inplace=True), #activation function
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        x = self.conv(x)
        return x


class inconv(nn.Module): #didnt understand how that class works
    def __init__(self, in_ch, out_ch):
        super(inconv, self).__init__()
        self.conv = double_conv(in_ch, out_ch)

    def forward(self, x):
        x = self.conv(x)
        return x


class down(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(down, self).__init__()
        self.mpconv = nn.Sequential(
            nn.MaxPool2d(2), #down-sample
            double_conv(in_ch, out_ch)
        )

    def forward(self, x):
        x = self.mpconv(x)
        return x


class up(nn.Module): #I don't understand well the upsampling part, how do we add those pixels? Are them from the "down part"?
    def __init__(self, in_ch, out_ch, bilinear=True):
        super(up, self).__init__()

        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        else:
            self.up = nn.ConvTranspose2d(in_ch//2, in_ch//2, 2, stride=2)

        self.conv = double_conv(in_ch, out_ch)

    def forward(self, x1, x2): #dont understand this part either
        x1 = self.up(x1)
        
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, (diffX // 2, diffX - diffX//2,
                        diffY // 2, diffY - diffY//2))

        x = torch.cat([x2, x1], dim=1)
        x = self.conv(x)
        return x

class outconv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(outconv, self).__init__()
        self.conv = nn.Conv2d(in_ch, out_ch, 1)

    def forward(self, x):
        x = self.conv(x)
        return x
```

```python
#import torch.nn.functional as F


class UNet(nn.Module):
    def __init__(self, n_channels, n_classes):
        super(UNet, self).__init__()
        self.inc = inconv(n_channels, 64)
        self.down1 = down(64, 128)
        self.down2 = down(128, 256)
        self.down3 = down(256, 512)
        self.down4 = down(512, 512)
        self.up1 = up(1024, 256)
        self.up2 = up(512, 128)
        self.up3 = up(256, 64)
        self.up4 = up(128, 64)
        self.outc = outconv(64, n_classes)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)
        x = self.outc(x)
        return torch.sigmoid(x)
```

```python
my_unet = UNet(n_channels = 3, n_classes=1)
```

```python
from tqdm import tqdm_notebook as tqdm
import torch.optim as optim

optimizer = optim.SGD(my_unet.parameters(), lr=0.001, momentum=0.9)
my_unet.cuda();
criterion = nn.BCELoss()
```

```python
my_unet.train();
```

```python
# Print unet's state_dict
#print("Unet's state_dict:")
# for param_tensor in my_unet.state_dict():
#     print(param_tensor, "\t", my_unet.state_dict()[param_tensor].size())

# Print optimizer's state_dict
#print("Optimizer's state_dict:")
# for var_name in optimizer.state_dict():
#     print(var_name, "\t", optimizer.state_dict()[var_name])



#Load the best model and optimizer, also handling the lack of save

lowest_loss = 10.0
epoch = 0

def load_checkpoint(model, optimizer):
        checkpoint = torch.load(PATH_TO_SAVE)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        return checkpoint

try:           
    checkpoint = load_checkpoint(my_unet, optimizer)
    lowest_loss = checkpoint['loss']
    epoch = checkpoint['epoch']
except OSError as e:
    print(e)        
```

```python
lowest_loss
```

```python
nr_epochs = 10
```

```python
for epoch in tqdm(range(nr_epochs)):   # loop over the dataset multiple times
    # train
    my_unet.train()
    running_loss = 0.0
    for i, batches in enumerate(my_loader):
        # get the inputs; data is a list of [inputs, labels]
        image_batch, labels_batch = batches
        image_batch, labels_batch  =image_batch.cuda(), labels_batch.cuda()
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs_batch = my_unet(image_batch)
        loss = criterion(outputs_batch, labels_batch)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 1 == 0:
            print('[%d, %5d] lossTrain: %.3f' %
                  (epoch + 1, i + 1, running_loss/my_loader.batch_size)) #better than hard coded denominator
            running_loss = 0.0
    # end of epoch
    
    # val
    my_unet.eval()
    for j, batches in enumerate(my_val_loader):
        # get the inputs; data is a list of [inputs, labels]
        image_batch, labels_batch = batches
        image_batch, labels_batch = image_batch.cuda(), labels_batch.cuda()

        # forward + NO BACKWARD / OPTIMIZE (validation, not training)
        outputs_batch = my_unet(image_batch)
        loss = criterion(outputs_batch, labels_batch) 
        
        #save if there is a new lowest loss
        if (loss < lowest_loss):
            lowest_loss = loss
            torch.save({
            'epoch': epoch,
            'model_state_dict': my_unet.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': loss
            }, PATH_TO_SAVE)
            
            
            print('%.3f is the new lowest loss' %loss)
                   
        # print statistics
        running_loss += loss.item()
        if j % 1 == 0:
            print('[%d, %5d] lossVal: %.3f' %
                  (epoch + 1, j + 1, running_loss/my_val_loader.batch_size)) #better than hard coded denominator
            running_loss = 0.0                        

    
print('Finished Training, the lowest loss is %.3f.' %lowest_loss)
```
